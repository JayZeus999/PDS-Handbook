```python
import re
```

```python
def clean_strings(strings): 
    result = [] 
    for value in strings: 
        value = value.strip() 
        value = re.sub('[!#?]', '', value) 
        value = value.title() 
        result.append(value)
    return result
```

```python
clean_strings(states)
```

```python
names = ['Alan', 'Adam', 'Wes', 'Will', 'Albert', 'Steven'] 
```

```python
for letter, group_names in itertools.groupby(names, first_letter):
    print(letter, list(group_names))  
    
# groupby only groups consecutive items, not all matching items.
# group_names → generator of items in that group
# groupby returns a generator, so you convert it to list to see values.
```
